- Part 3 
Data Balancing

The best approach is to combine undersampling technique and follow for oversampling technique,

```python
import imblearn
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

- First we used random undersampling()  We used Smapling Strategy= 0.1 to  make the minority class count 10% of the majority class count
- And then For oversampling we used SMOTE (Synthetic Minority Oversampling technique) We used Smapling Strategy= 0.5 to make the minority class count 50% of the majority class count


```python
over = SMOTE(sampling_strategy = 0.5)
under = RandomUnderSampler(sampling_strategy = 0.1)
X = df1.iloc[:,:9].values
Y = df1.iloc[:,9].values


pipeline = Pipeline(steps= [('under', under),('over', over)])
X, Y = pipeline.fit_resample(X, Y)
y = pd.DataFrame(Y)
y.value_counts()
```

- Part 5 Modeling

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state = 42)
```

```python

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier

knn =KNeighborsClassifier(leaf_size = 1, n_neighbors = 3,p = 1)

knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
cv = RepeatedStratifiedKFold(n_splits = 10,n_repeats = 3,random_state = 42)
print("Cross Validation Score : ",'{0:.2%}'.format(cross_val_score(knn,x_train,y_train,cv = cv,scoring = 'roc_auc').mean()))
print("ROC_AUC Score : ",'{0:.2%}'.format(roc_auc_score(y_test,y_pred)))



# Confusion Matrix
cm = confusion_matrix(y_test,y_pred)
names = ['True Neg','False Pos','False Neg','True Pos']
counts = [value for value in cm.flatten()]
percentages = ['{0:.2%}'.format(value) for value in cm.flatten()/np.sum(cm)]
labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(names,counts,percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm,annot = labels,cmap = 'Blues',fmt ='')

# Classification Report
print(classification_report(y_test,y_pred))
```

```python
import pickle

pickle.dump(knn,open('knn_model.pkl','wb'))
```
